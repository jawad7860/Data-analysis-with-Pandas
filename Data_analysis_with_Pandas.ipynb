**1. How many men and women (*sex* feature) are represented in this dataset?** 

```python
sex_feature=data['sex']

m_sum=0
f_sum=0

for i in sex_feature:

    if i=="Male":
        
        m_sum=m_sum+1
    
    elif i=="Female":
        f_sum=f_sum+1

print("Male  ",m_sum)
print("Female  ",f_sum)

data['sex'].value_counts()
```

**2. What is the average age (*age* feature) of women?**

```python
F_Rows=data[data['sex']=='Female']

print(F_Rows['age'].mean())
```

**3. What is the proportion of German citizens (*native-country* feature)?**

```python
German=data[data['native-country']=='Germany']

Total_Germans=German['native-country'].count()

Total_Count=data['native-country'].count()

print(Total_Germans/Total_Count)


#data.info() #32561
```

**4-5. What are mean value and standard deviation of the age of those who recieve more than 50K per year (*salary* feature) and those who receive less than 50K per year? **

```python
 

More=data[data['salary']=='>50K']

Less=data[data['salary']=='<=50K']

Rich_age=round(More['age'].mean())

Poor_age=round(Less['age'].mean())

Poor_std=Less['age'].std()

Rich_std=More['age'].std()

#print(Poor_age,Poor_std)

print("The average age of rich: ",Rich_age," +- %.1f "%Rich_std," Poor: ",Poor_age," +- %.1f "%Poor_std)
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours and what is the percentage of those who earn a lot among them?**

```python

hours=data['hours-per-week'].max()

people=data[data['hours-per-week']==hours]

people=people.shape[0]

Rich=More[More['hours-per-week']==hours]

Rich=Rich.shape[0]

percentage=(Rich/people)*100

print("Max time - ",hours," hours./week.")
print("Total number of such hard workers ",people)
print("Percentage of rich among them ",percentage)
```
